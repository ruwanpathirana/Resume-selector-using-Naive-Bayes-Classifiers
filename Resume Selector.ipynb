## PERFORM DATA CLEANING

```python
# remove back slash
resume_df['resume_text'] = resume_df['resume_text'].apply(lambda x:x.replace('\r', ''))
resume_df
```

```python
# download nltk packages
nltk.download('punkt')
```

```python
# download nltk packages
nltk.download("stopwords")
```

```python
# Get additional stopwords from nltk
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['from','subject','re','edu','use','email','com'])
```

```python
# Remove stop words and remove words with 2 or less characters because thats are didn't effect for resumes
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text) :
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 2 and token not in stop_words:
            result.append(token)
            
    return ' '.join(result)
```

```python
# Cleaned text
resume_df['cleaned'] = resume_df['resume_text'].apply(preprocess)
```

## TRAIN NAIVE BAYES CLASSIFIER MODEL

```python
countvectorizer
```

```python
X = countvectorizer
y = resume_df['class']
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

```python
X_train.shape
```

```python
X_test.shape
```

## ASSESS TRAINED MODEL PERFORMANCE


```python
# Predicting the performance on train data (Only for comparing with test prediction)
y_predict_train = NB_classifier.predict(X_train)
y_predict_train
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot = True)
```

```python
# Predicting the Test set results
y_predict_test = NB_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot = True)
```

```python
# classification report
print(classification_report(y_test, y_predict_test))
```
